```python
from sklearn.ensemble import StackingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.svm import SVR
```

#### Creating objects of the models

```python
rfr=RandomForestRegressor(random_state=42)
dtr=DecisionTreeRegressor(random_state=42)
gbr=GradientBoostingRegressor(random_state=42)
adr=AdaBoostRegressor(random_state=42)
svr=SVR()
```

### **Creating Base learners**

```python
estimators=[
    ('RandForest',rfr),
    ('Support Vector Regressor',svr),
    ('Gradient Boosting Regressor',gbr),
    ('AdaBoost Regressor',adr),
    ('DecisionTreeRegressor',dtr)
]
```

### <font color='green'>Creating the Meta Model

```python
lr=LinearRegression()
```

#### **Combining the base learners and meta model so that predictions from the base learners are given as cross validated input to meta model, reducing overfitting**

```python
stacking_reg=StackingRegressor(estimators=estimators,final_estimator=lr,cv=5)
```

#### **Training Stacking Classifier**

```python
stacking_reg.fit(X_train,y_train)
```

#### Collecting the r2_score from each base Learner

```python
models = []
scores = []
losses = []
for model in estimators:
    models.append(model[0])
    model[1].fit(X_train, y_train)
    y_pred_s = model[1].predict(X_test)
    score = r2_score(y_test, y_pred_s)  # Calculate RÂ² score
    scores.append(score)  # Append the score to the list # Print predictions
    losses.append(root_mean_squared_error(y_test,y_pred_s))
    print(f"{model[0]} RÂ² score: {score}")  # Print the model name and score
```

#### Generating Predictions from the Stacking Classifier

```python
y_pred_stack=stacking_reg.predict(X_test)
```

## **Evaluating the Stacking Classifier**

```python
r2_score_stack=r2_score(y_test,y_pred_stack)
mae_stack=mean_absolute_error(y_test,y_pred_stack)
rmse_stack=root_mean_squared_error(y_test,y_pred_stack)
models.append("Stacking Classifier")
scores.append(r2_score_stack)
losses.append(rmse_stack)
```

***Displaying Results***

```python
print(f"R2-score of Stacking Regression::{r2_score_stack}")
print(f"Mean Absolute error of Stacking Regression::{mae_stack}")
print(f"Root Mean Squared Error of Stacking Regression::{rmse_stack}")
```

```python
bars = plt.bar(x=models, height=scores, color=['red', 'green', 'blue', 'brown', 'teal', 'purple'])
plt.xticks(rotation=90)
plt.ylabel("R2_Score", weight='bold')
plt.title("Comparison of R2-Score between Different Models")
plt.bar_label(bars)
# Show the plot
plt.show()
```

```python
bars = plt.bar(x=models, height=losses, color=['red', 'green', 'blue', 'brown', 'teal', 'purple'])
plt.xticks(rotation=90)
plt.ylabel("RMSE", weight='bold')
plt.title("Comparison of RMSE between Different Models")
plt.bar_label(bars)
# Show the plot
plt.show()
```
